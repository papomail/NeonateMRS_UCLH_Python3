```python
import numpy as np
import matplotlib.pyplot as plt
#import scipy.fftpack as fft
import numpy.fft as fft
from math import pi
```

```python
Fs=1000

t=np.arange(0,1,1/Fs)
ph1=np.pi/2

f1=10
a1=3
sig1=a1*( np.cos(2*pi*f1*t+ph1) + 1j*np.sin(2*pi*f1*t+ph1)  )



f2=40
a2=2
sig2=a2*( np.cos(2*pi*f2*t) + 1j*np.sin(2*pi*f2*t)  )

f3=50
a3=4
ph3=0*np.pi

sig3=a3*np.exp(1j*(2*np.pi*f3*t+ph3))

kspace=sig1+sig2+sig3
plt.subplot(2,1,1); plt.plot(t,np.real(kspace)); plt.ylabel('Real')
plt.subplot(2,1,2); plt.plot(t,np.imag(kspace)); plt.ylabel('Imaginary')


frq=Fs/2*np.linspace(0,1,len(t)/2)


FT=(fft.fft(kspace))

spect=2/len(t)*FT[0:len(frq)]


frq2=fft.fftfreq(1000)
spect2=2/len(t)*FT
```

```python
plt.subplot(2,1,1)
plt.plot(t,np.abs(kspace)); plt.title('Test signal') 
plt.ylabel('Amplitude'); plt.xlabel('Time (s)')

plt.subplot(2,1,2)
plt.plot(frq2,np.abs(spect2)**2); plt.title('Spectrum')
plt.ylabel('Power spectrum'); plt.xlabel('Frequency (Hz)')

plt.tight_layout()
```
